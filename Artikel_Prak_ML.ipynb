```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
import time

print("Memuat dataset CIFAR-10...")
(X_train_raw, y_train_raw), (X_test_raw, y_test_raw) = cifar10.load_data()

X = np.concatenate((X_train_raw, X_test_raw), axis=0)
y = np.concatenate((y_train_raw, y_test_raw), axis=0)

fraction_size = 0.1
np.random.seed(42)
subset_indices = np.random.choice(len(X), int(len(X) * fraction_size), replace=False)
X_subset = X[subset_indices]
y_subset = y[subset_indices]

X_train, X_test, y_train, y_test = train_test_split(
    X_subset, y_subset, test_size=0.2, random_state=42, stratify=y_subset
)

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

print(f"Bentuk data pelatihan (gambar): {X_train.shape}")
print(f"Bentuk label pelatihan: {y_train.shape}")
print(f"Bentuk data pengujian (gambar): {X_test.shape}")
print(f"Bentuk label pengujian: {y_test.shape}")



X_train_flatten = X_train.reshape(X_train.shape[0], -1).astype('float32') / 255.0
X_test_flatten = X_test.reshape(X_test.shape[0], -1).astype('float32') / 255.0

X_train_cnn = X_train.astype('float32') / 255.0
X_test_cnn = X_test.astype('float32') / 255.0

num_classes = len(np.unique(y_train))
y_train_cnn = to_categorical(y_train, num_classes)
y_test_cnn = to_categorical(y_test, num_classes)

results = {}

print("\n--- Melatih K-Nearest Neighbors (KNN) ---")
start_time_knn = time.time()
scaler = StandardScaler()
X_train_scaled_knn = scaler.fit_transform(X_train_flatten)
X_test_scaled_knn = scaler.transform(X_test_flatten)

knn_model = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
knn_model.fit(X_train_scaled_knn, y_train.ravel())
y_pred_knn = knn_model.predict(X_test_scaled_knn)
end_time_knn = time.time()

results['KNN'] = {
    'accuracy': accuracy_score(y_test, y_pred_knn),
    'report': classification_report(y_test, y_pred_knn, target_names=class_names, zero_division=0),
    'confusion_matrix': confusion_matrix(y_test, y_pred_knn),
    'training_time': 0,
    'prediction_time': end_time_knn - start_time_knn
}
print(f"Akurasi KNN: {results['KNN']['accuracy']:.4f}")
print(f"Waktu Komputasi KNN (termasuk prediksi): {results['KNN']['prediction_time']:.2f} detik")


print("\n--- Melatih Support Vector Machine (SVM) ---")
start_time_svm = time.time()
svm_model = SVC(kernel='linear', random_state=42, verbose=False)
svm_model.fit(X_train_scaled_knn, y_train.ravel())
end_time_svm_train = time.time()
y_pred_svm = svm_model.predict(X_test_scaled_knn)
end_time_svm_total = time.time()

results['SVM'] = {
    'accuracy': accuracy_score(y_test, y_pred_svm),
    'report': classification_report(y_test, y_pred_svm, target_names=class_names, zero_division=0),
    'confusion_matrix': confusion_matrix(y_test, y_pred_svm),
    'training_time': end_time_svm_train - start_time_svm,
    'prediction_time': end_time_svm_total - end_time_svm_train
}
print(f"Akurasi SVM: {results['SVM']['accuracy']:.4f}")
print(f"Waktu Pelatihan SVM: {results['SVM']['training_time']:.2f} detik")
print(f"Waktu Prediksi SVM: {results['SVM']['prediction_time']:.2f} detik")


print("\n--- Melatih Convolutional Neural Network (CNN) ---")
start_time_cnn = time.time()
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=X_train_cnn.shape[1:]),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])

cnn_model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

cnn_model.summary()

history = cnn_model.fit(X_train_cnn, y_train_cnn,
                        epochs=10,
                        batch_size=64,
                        validation_data=(X_test_cnn, y_test_cnn),
                        verbose=1)
end_time_cnn_train = time.time()

loss_cnn, accuracy_cnn = cnn_model.evaluate(X_test_cnn, y_test_cnn, verbose=0)
y_pred_cnn_probs = cnn_model.predict(X_test_cnn)
y_pred_cnn = np.argmax(y_pred_cnn_probs, axis=1)
end_time_cnn_total = time.time()

results['CNN'] = {
    'accuracy': accuracy_cnn,
    'report': classification_report(np.argmax(y_test_cnn, axis=1), y_pred_cnn, target_names=class_names, zero_division=0),
    'confusion_matrix': confusion_matrix(np.argmax(y_test_cnn, axis=1), y_pred_cnn),
    'training_time': end_time_cnn_train - start_time_cnn,
    'prediction_time': end_time_cnn_total - end_time_cnn_train
}
print(f"Akurasi CNN: {results['CNN']['accuracy']:.4f}")
print(f"Waktu Pelatihan CNN: {results['CNN']['training_time']:.2f} detik")
print(f"Waktu Prediksi CNN: {results['CNN']['prediction_time']:.2f} detik")

print("\n--- Ringkasan Hasil Komparasi Akurasi dan Waktu ---")
print("{:<15} {:<10} {:<18} {:<18}".format("Algoritma", "Akurasi", "Waktu Pelatihan (s)", "Waktu Prediksi (s)"))
print("-" * 61)
for algo, res in results.items():
    train_time_str = f"{res['training_time']:.2f}" if 'training_time' in res else "N/A"
    pred_time_str = f"{res['prediction_time']:.2f}" if 'prediction_time' in res else "N/A"
    print(f"{algo:<15} {res['accuracy']:.4f} {train_time_str:<18} {pred_time_str:<18}")


algorithms = list(results.keys())
accuracies = [results[algo]['accuracy'] for algo in algorithms]

plt.figure(figsize=(8, 6))
plt.bar(algorithms, accuracies, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel("Algoritma", fontsize=12)
plt.ylabel("Akurasi", fontsize=12)
plt.title("Perbandingan Akurasi Klasifikasi Gambar pada CIFAR-10", fontsize=14)
plt.ylim(0, 1)
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.02, f'{acc:.4f}', ha='center', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

training_times = [results[algo]['training_time'] if 'training_time' in results[algo] else 0 for algo in algorithms]
prediction_times = [results[algo]['prediction_time'] if 'prediction_time' in results[algo] else results[algo]['training_time'] for algo in algorithms] # KNN has prediction time as total time

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Algoritma', fontsize=12)
ax1.set_ylabel('Waktu Pelatihan (Detik)', color=color, fontsize=12)
bar_width = 0.35
x = np.arange(len(algorithms))

rects1 = ax1.bar(x - bar_width/2, training_times, bar_width, label='Waktu Pelatihan', color='lightcoral', alpha=0.8)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yscale('log')
ax1.set_ylim(bottom=0.01)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Waktu Prediksi (Detik)', color=color, fontsize=12)
rects2 = ax2.bar(x + bar_width/2, prediction_times, bar_width, label='Waktu Prediksi', color='lightgreen', alpha=0.8)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yscale('log')
ax2.set_ylim(bottom=0.01)

plt.xticks(x, algorithms)
ax1.set_title('Perbandingan Waktu Pelatihan dan Prediksi pada CIFAR-10', fontsize=14)
fig.tight_layout()
fig.legend(loc="upper center", bbox_to_anchor=(0.5, 1.05), ncol=2, fontsize=10)
plt.show()


print("\n--- Visualisasi Confusion Matrix ---")
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes = axes.flatten()

for i, (algo, res) in enumerate(results.items()):
    sns.heatmap(res['confusion_matrix'], annot=True, fmt='d', cmap='Blues', ax=axes[i], cbar=False,
                xticklabels=class_names, yticklabels=class_names)
    axes[i].set_title(f'Confusion Matrix - {algo}', fontsize=14)
    axes[i].set_xlabel('Prediksi', fontsize=10)
    axes[i].set_ylabel('Aktual', fontsize=10)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].tick_params(axis='y', rotation=0)

plt.tight_layout()
plt.show()

print("\n--- Laporan Klasifikasi Lengkap per Algoritma ---")
for algo, res in results.items():
    print(f"\n### {algo} ###")
    print(res['report'])
```
